```python
#upper fence and lower fence
q1=df.imdb_votes.quantile(0.25)
med=df.imdb_votes.median()
q3=df.imdb_votes.quantile(0.75)
iqr=q3-q1
print('IQR\t\t',q3-q1)
uf=q3+1.5*iqr
lf=q1-1.5*iqr
print('upper fence\t',uf)
print('lower fence\t',lf)
print('numbers of outliers\t',df[df.imdb_votes>uf].shape[0])
print('potential outliers')
df[df.imdb_votes>uf].sample(5)
```
